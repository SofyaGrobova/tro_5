Выбираем данные по количеству детей и строим диаграмму

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv ('st.csv')
```

```python
ch = df['children'].value_counts()
keys = list(dict(ch).keys())
values = list(dict(ch).values())

fig, ax = plt.subplots()
ax.pie(values, labels=keys, autopct='%1.1f%%')
ax.axis("equal")
```

Из диаграммы видно, что количество регионов в которых люди получили страховки примерно равно.

Выбираем данные по полу, людей которые у которых есть дети и строим диаграмму

```python
s = df['sex'].value_counts()

keys = list(dict(s).keys())
values = list(dict(s).values())

fig, ax = plt.subplots()
ax.pie(values, labels=keys, autopct='%1.1f%%')
ax.axis("equal")
```
